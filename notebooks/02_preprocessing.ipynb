```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

X_train_processed.shape, X_test_processed.shape
```

```python
feature_names = preprocessor.get_feature_names_out()
feature_names
```

```python
import pandas as pd

X_train_processed_df = pd.DataFrame(
    X_train_processed,
    columns=feature_names,
    index=X_train.index
)

X_test_processed_df = pd.DataFrame(
    X_test_processed,
    columns=feature_names,
    index=X_test.index
)
```

```python
from pathlib import Path

processed_path = Path("../data/processed")
processed_path.mkdir(parents=True, exist_ok=True)
```

```python
X_train_processed_df.to_csv(processed_path / "X_train.csv", index=False)
X_test_processed_df.to_csv(processed_path / "X_test.csv", index=False)
```

```python
y_train.to_csv(processed_path / "y_train.csv", index=False)
y_test.to_csv(processed_path / "y_test.csv", index=False)
```

```python
import json

metadata = {
    "numerical_features": numerical_features,
    "categorical_features": categorical_features,
    "numerical_imputation": "median",
    "categorical_imputation": "most_frequent",
    "scaling": "StandardScaler",
    "encoding": "OneHotEncoder",
    "split": "80/20 stratified",
    "output_shape_train": X_train_processed_df.shape,
    "output_shape_test": X_test_processed_df.shape
}

with open(processed_path / "metadata.json", "w") as f:
    json.dump(metadata, f, indent=4)
```
